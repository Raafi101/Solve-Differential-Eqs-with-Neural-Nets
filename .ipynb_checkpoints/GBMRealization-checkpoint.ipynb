```python
from tqdm import tqdm
import numpy as np
import math
from scipy import stats
from matplotlib import pyplot as plt
```

```python
np.random.seed(301)
```

```python
def f(x,mu,sigma):
    f = mu*x
    return f

def g(x,mu,sigma):
    g = sigma*x
    return g
```

```python
t = np.linspace(0,1,10000)
```

```python
tf = t[-1]
dt = t[1]-t[0]
n = len(t)
r = np.zeros(1)
Y = np.zeros(n)
K1 = np.zeros(n)
```

```python
Y0 = 1 #initial condition
Y[0] = Y0
pi = math.pi
mu = 0
sigma = (1/32)
```

```python
for i in tqdm(range(n-1)):
    r = np.random.normal(0,1,1)
    dW = (r*np.sqrt(dt))
    #Yn = fdt + gdW
    K1[i] = f(Y[i],mu,sigma)*dt + g(Y[i],mu,sigma)*dW;
    Y[i+1] = Y[i] + K1[i]
```

```python
plt.plot(t,Y)
plt.xlabel('t')
plt.ylabel('X')
plt.title('Realization of Geometric Brownian Motion')
```
